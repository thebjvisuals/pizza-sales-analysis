```python
#data cleaning and preparation

#converting date/time

df["date"] = pd.to_datetime(df["date"])
df["date"]
```

```python
df["time"] = pd.to_datetime(df["time"], format="%H:%M:%S").dt.time
df["time"]
```

```python
#creating revenue column

df["revenue"] = df["quantity"] * df["price"]
df["revenue"]
```

```python
#total revenue

total_revenue = df["revenue"].sum()
print(total_revenue)
```

```python
#total quantity sold

total_quantity = df["quantity"].sum()
print(total_quantity)
```

```python
#total number of orders

total_orders = df["order_id"].nunique()
print(total_orders)
```

```python
#number of pizza types

pizza_types_count = df["name"].nunique()
print(pizza_types_count)
```

```python
#average pizza price

avg_price = df["price"].mean()
print(avg_price)
```

```python
#peak sales hours

df["hour"] = pd.to_datetime(df["time"], format="%H:%M:%S").dt.hour
df["hour"]
```

```python
peak_hours = df.groupby("hour")["revenue"].sum()
print(peak_hours)
```

```python
#plotting

peak_hours.plot(kind="bar")
plt.title("Revenue by Hour")
plt.show()
```

```python
#sales by day of week

df["day"] = df["date"].dt.day_name()
sales_by_day = df.groupby("day")["revenue"].sum().sort_values()
print(sales_by_day)
```

```python
#top 5 best-selling pizzas

top5 = df.groupby("name")["quantity"].sum().sort_values(ascending=False).head()
print(top5)
```

```python
#monthly sales trend

df["month"] = df["date"].dt.month_name()
monthly_sales = df.groupby("month")["revenue"].sum()
print(monthly_sales)
```

```python
#plot trend

monthly_sales.plot(kind="line", marker="o")
plt.title("Monthly Revenue Trend")
plt.show()
```

```python
#poor-performing pizzas

low_sales = df.groupby("name")["quantity"].sum().sort_values().head()
print(low_sales)
```

```python
#revenue by pizza category

df.groupby("category")["revenue"].sum().plot(kind="bar")
df
```

## 3. Sales Trend Analysis

### Peak Sales Hours

Sales are highest around **12 hour(s)** of the day.

This suggests that customer traffic is strongest during these hours, likely corresponding to lunch or dinner periods.

### Sales by Day of the Week

The highest revenue was generated on **Friday**.

This indicates that this day experiences the strongest customer turnout and may require increased staffing or inventory.

### Monthly Sales Trend

Sales peaked in **July** and were lowest in **October**.

This may indicates seasonal trends or promotional impacts affecting customer purchasing behavior.
